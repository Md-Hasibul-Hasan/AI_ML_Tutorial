```python
def ep(txt):
    data = ''
    inside = False
    for i in txt:
        if i=='(':
            inside = True
            continue
        if i == ')':
            break

        if inside:
            data+=i
    return data
```

```python
df['Episodes']=df['Title'].apply(ep)
df['Episodes']=df['Episodes'].str.replace(' eps','')
df['Episodes']=df['Episodes'].astype(int)
df.loc[0]['Episodes']
```

```python
df.loc[1]['Title']
```

```python
def time_period(txt):
    inside = False
    data = ''
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1,i+20):
                data+=txt[j]
            return data

# time_period('Steins;GateTV (24 eps)Apr 2011 - Sep 20112,473,707 members')
```

```python
df['Time Period']=df['Title'].apply(time_period)
df[:5]
```

Calculate Total Month

```python


def total_months(period):
    start, end = period.split(' - ')

    start = pd.to_datetime(start, format='%b %Y')
    end = pd.to_datetime(end, format='%b %Y')

    return (end.year - start.year) * 12 + (end.month - start.month)

total_months('Apr 2009 - Jul 2010')
```

```python
df['Months'] = df['Time Period'].apply(total_months)
df[:5]
```
