```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sns
```

# Matplot

```python
x = np.linspace(0,10,20)
y = x**3
mat.plot(x,y)
mat.title("Cubic Function")
mat.xlabel("X-axis")
mat.ylabel("Y-axis")
```

```python
a = np.arange(-10, 11)
b = a**2

# mat.plot(a, -b)
mat.plot(a, -b,
         marker='o',
         markersize=3)

for i, j in zip(a, -b):
    mat.text(i, j, f'({i},{j})')

mat.xlim(-15, 15)
mat.ylim(-110, 110)

mat.axhline(0, color='black')
mat.axvline(0, color='black')

mat.grid(True)

mat.show()
```

```python
#subplot in matplotlib less flexible than object oriented way but it is easier to use
mat.subplot(2,2,1)
mat.plot(x,y)
mat.subplot(2,2,2)
mat.plot(y,x)
mat.subplot(2,2,3)
mat.plot(x,x**2)
mat.subplot(2,2,4)
mat.plot(x,x**3)
```

```python
# Object Oriented Way
fig = mat.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y)
ax2 = fig.add_axes([0.4,0.6,0.3,0.3])
ax2.plot(y,x)
```

```python
fig = mat.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,y)
ax1.set_title("Cubic Function")
ax1.set_ylabel("Y-axis")
ax1.set_xlabel("X-axis")

ax2 = fig.add_subplot(2,2,2)
ax2.plot(y,x)
ax3 = fig.add_subplot(2,2,3)
ax3.plot(x,x**2)
ax4 = fig.add_subplot(2,2,4)
ax4.plot(x,x**3)


mat.tight_layout()
mat.show()
```

```python
fig, axes = mat.subplots(2, 2)

axes[0,0].plot(x,y)
axes[0,0].set_title("Cubic Function")
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

axes[0,1].plot(y,x)
axes[1,0].plot(x,x**2)
axes[1,1].plot(x,x**3)

mat.tight_layout()
mat.show()
```

# Distribution Plot: displot/hisplt, jointplot, pairplot

```python
df['total_bill'].hist() # by pandas method
```

```python
mat.hist(df['total_bill']) # by matplotlib method, it is the same as above but more flexible to customize
```

```python
sns.displot(df['total_bill'], bins=30, kde=True) # by seaborn method
```

```python
mat.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=30, kde=True)
mat.title("Total Bill Distribution")

mat.subplot(1,2,2)
sns.histplot(df['tip'], bins=30, kde=True, color='orange', edgecolor='blue')
mat.title("Tips Distribution")
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter',)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

# Matrix Plot : Heat map, Cluster Map, Pivot table heat map

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
# correlation 
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

```python
sns.clustermap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

```python
pvtflights = flights.pivot_table(index='month', columns='year', values='passengers')
pvtflights
```

```python
sns.heatmap(pvtflights)
```
